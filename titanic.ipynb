```python
import pandas as pd
```

```python
df = pd.read_csv("/content/titanic.csv")
df.head()
```

```python
dfd = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 'columns')
dfd
```

```python
dfd.fillna(method = 'ffill',inplace = True)
dfd
```

```python
input_n = dfd.drop('Survived', axis = 'columns')
target = dfd.Survived
input_n
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
input_n['Sex_n'] = le.fit_transform(input_n['Sex'])
input_n.head()
```

```python
final = input_n.drop('Sex',axis = 'columns')
final
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
x = final
y = target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[2,26,14,1]])
```
